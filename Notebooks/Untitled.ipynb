```python
documents = []
for i, x in enumerate(docs):
    print(i, x)
    row ={
        'id': i,
        'text': x
    }
    documents.append(row)
documents
```

```python
datasets = [{'id': 0, 'text': 'The weather is lovely today.'},
            {'id': 1, 'text': "It's so sunny outside!"},
            {'id': 2, 'text': 'He drove to the stadium.'},
            {'id': 3, 'text': 'the dog is green'}]
```

```python
documents_embeddings = []
for x in datasets:
    embed = model.encode(x.get('text'))
    final_data = (x.get('id'),embed)
    documents_embeddings.append(final_data)
documents_embeddings[0]
```

```python
query = "the dog is funny and jumping in the rain"
query_embedding = model.encode(query)
```

```python
results = []
for doc in documents_embeddings:
    doc_id = doc[0]
    doc_embedding = doc[1]
    rank = model.similarity(doc_embedding, query_embedding)
    results.append(
        (doc_id, rank)
    )
results
```

```python
results.sort(key=lambda x: x[1], reverse=True)
```

```python
for result in results:
    dataset_id = result[0]
    rank = result[1]
    doc = next(doc for doc in datasets if doc['id'] == dataset_id)
    print(dataset_id, rank, doc)
```
